### Creating lists and appending datas related to BeatifulSoup (For BMW)

```python
count = 0
headline = list()
seher = list()
model = list()
buraxilis_ili = list()
ban_novu = list()
color = list()
motor = list()
km = list()
manual_or_auto = list()
oturucu = list()
new_or_old = list()
kredit_ve_barter = list()
shop_owner = list()
price = list()
extras = list()
offer_from_abroad = list()
image = list()
yerlerin_sayi = list()
veziyyeti = list()
bazar = list()
product_urls = list()


for url in  tqdm.tqdm(all_turbo_urls):

    product_urls.append(url)

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.93 Safari/537.36',
    }

    sent_request = requests.get(url, headers= headers)
    
    nonsense = True
    while nonsense:
        if sent_request.status_code != 200:
            
            time.sleep(1.2)
            sent_request = requests.get(url, headers= headers)
            count += 1
            
            
            if count == 5:
                print(f'Request for {url} has been unsucessful!')
                break
            
        elif sent_request.status_code == 200:
            nonsense = False
        
    
        
    web_page_html_code = sent_request.content
    bs = BeautifulSoup(web_page_html_code, 'html.parser')
    
    try:
        baslig = bs.find('h1', {'class': 'product-title'}).text
        headline.append(baslig)
    except:
        headline.append(None)

    try:
        seherler = bs.find_all('span', {'class': 'product-properties__i-value'})[0].text
        seher.append(seherler)
    except:
        seher.append(None)

    try:
        modeller = bs.find_all('span', {'class': 'product-properties__i-value'})[2].text
        model.append(modeller)
    except:
        model.append(None)

    try:
        burax_il = bs.find_all('span', {'class': 'product-properties__i-value'})[3].text
        buraxilis_ili.append(burax_il)
    except:
        buraxilis_ili.append(None)

    try:
        ban = bs.find_all('span', {'class': 'product-properties__i-value'})[4].text
        ban_novu.append(ban)
    except:
        ban_novu.append(None)

    try:
        reng = bs.find_all('span', {'class': 'product-properties__i-value'})[5].text
        color.append(reng)
    except:
        color.append(None)

    try:
        muherrik = bs.find_all('span', {'class': 'product-properties__i-value'})[6].text
        motor.append(muherrik)
    except:
        motor.append(None)

    try:
        yuruyus = bs.find_all('span', {'class': 'product-properties__i-value'})[7].text
        km.append(yuruyus)
    except:
        km.append(None)
    
    try:
        suretler_qutusu = bs.find_all('span', {'class': 'product-properties__i-value'})[8].text
        manual_or_auto.append(suretler_qutusu)
    except:
        manual_or_auto.append(None)

    try:
        oturucu_tipi = bs.find_all('span', {'class': 'product-properties__i-value'})[9].text
        oturucu.append(oturucu_tipi)
    except:
        oturucu.append(None)

    try:
        yeni_kohne = bs.find_all('span', {'class': 'product-properties__i-value'})[10].text
        new_or_old.append(yeni_kohne)
    except:
        new_or_old.append(None)
        
    try:
        labels = bs.find_all('div', {'class': 'product-labels__i tz-d-flex tz-align-center'})
        
        indices_to_extract = [0, 1]

        results = [
            labels[i].text.strip() if i < len(labels) else 'None'
            for i in indices_to_extract
        ]
        
        kredit_barter = ','.join(results)
        kredit_ve_barter.append(kredit_barter)

    except:
        pass

    try:
        resmi_numayende = bs.find('div', {'class': 'product-shop__owner-name'}).text
        shop_owner.append(resmi_numayende)
    except:
        shop_owner.append(None)

    try:
        qiymet = bs.find('div', {'class': 'product-price__i product-price__i--bold'}).text
        price.append(qiymet)
    except:
        price.append(None)

    try:
        merging = [i.text.strip() for i in bs.find_all('li', class_='product-extras__i')]
        
        if merging:
            extras.append(', '.join(merging))
        else:
            extras.append(None)
    except:
        extras.append(None)

    try:
        offer = bs.find('div', {'class': 'product-shop__status_order'}).text
        offer_from_abroad.append(offer)
    except:
        offer_from_abroad.append(None)

    try:
        sekil = bs.find_all('span', {'class': 'product-photos__slider-top-i_background'})[0]

        image_string = str(sekil)

        image_before_url = image_string.split("url('")[1]
        image_after_url = image_before_url.split("')")[0]

        image.append(image_after_url)
    except:
        image.append(None)

    try:
        properties_dict = {}
        property_blocks = bs.find_all('div', {'class': 'product-properties__i'})
        
        for block in property_blocks:
            name_tag = block.find('label', {'class': 'product-properties__i-name'})
            value_tag = block.find('span', {'class': 'product-properties__i-value'})
            if name_tag and value_tag:
                label = name_tag.text.strip()
                value = value_tag.text.strip()
                properties_dict[label] = value

        yerlerin_sayi.append(properties_dict.get('Yerlərin sayı', None))
        veziyyeti.append(properties_dict.get('Vəziyyəti', None))
        bazar.append(properties_dict.get('Hansı bazar üçün yığılıb', None))
    except:
        yerlerin_sayi.append(None)
        veziyyeti.append(None)
        bazar.append(None)
```

```python
bmw_df = pd.DataFrame(zip(headline, seher, model, buraxilis_ili, ban_novu, 
                 color, motor, km, manual_or_auto, oturucu, new_or_old, 
                     kredit_ve_barter, shop_owner, price, extras, offer_from_abroad, image, 
                          yerlerin_sayi, veziyyeti, bazar, product_urls) ,
                      
            columns= ['Bashlig', 'Sheher', 'Model', 'Buraxilish ili', 'Ban novu',
                      'Reng', 'Motor', 'Km', 'Suretler qutusu', 'Oturucu', 'Yeni ya Kohne', 
                     'Kredit ve Barter', 'Resmi Satish', 'Qiymet', 'Elaveler', 'Sifarisle', 'Shekil urli', 
                      'Yerlerin sayi', 'Veziyyeti', 'Hansi Bazar', 'Url'])
```

```python
bmw_df.to_excel('BMW_new.xlsx', index=False)
```
